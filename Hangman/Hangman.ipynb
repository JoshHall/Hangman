```python
## Import statements
import random 
from IPython.display import clear_output

# TODO, Create the class to hold the methods and attributes
class Hangman():
    # TODO, Create the __init__ and any attributes wanted
    code_words = [['Code Terms','method'],['Code Terms','function'],['Code Terms','class'],['Code Terms','variable'],
             ['Code Terms','list'],['Code Terms','dictionary'],['Code Terms','string'],['Code Terms','elif'],
             ['Code Terms','while loop'],['Code Terms',' for loop'],['Code Terms','scope'],['Code Terms','tuple'],
             ['Code Terms','module'],['Code Terms','frozen set']]
    sports_words = [['Sports','baseball'],['Sports','basketball'],
             ['Sports','soccer'],['Sports','football'],['Sports','hockey'],['Sports','cricket'],['Sports','rugby'],
             ['Sports','snowboarding'],['Sports','skiing'],['Sports','wake boarding'],['Sports','water skiing'],
             ['Sports','skateboarding'],['Sports','shooting']]
    
    def __init__(self,lives = 7, flag = True):
        self.lives = lives
        self.random_word = ''
        self.board = ''
        self.ans = ''
        self.ansL = ''
        self.ansU = ''
        self.flag = flag
        self.toBe = []
        self.guesses = ''
        pass
    
    # TODO, Create a method for picking a random word
    def randomWord(self,ans):
        self.ans = ans
        if ans == '1' or ans == 'code terms':
            toBoard = random.choice(self.code_words)
            toBoard = toBoard[1]
            return toBoard
        elif ans == '2' or ans == 'sports':
            toBoard = random.choice(self.sports_words)
            toBoard = toBoard[1]
            return toBoard
        else:
            print('Please enter an option give.')
            
            
        
    # TODO, Create a method for making a board
    def makeBoard(self,ans):
        randomWord = self.randomWord(ans)
        randWord = randomWord.title()
        for l in randWord:
            if l == ' ':
                self.toBe.append(' ')
            self.toBe.append('_ ')
        
        board = ''
        for l in self.toBe:
            if l == ' ':
                board += ' '
            board += '_ '
        
        self.board = board
        print(board)
        # Store the random word
        self.random_word = randWord
#         print(self.random_word)
        self.makeGuess()
        self.printMan()
    
    def printMan(self):
        d = {
            'p1' : '       _______',
            'p2' : '       |      |',
            'p3' : '              |',
            'p4' : '              |',
            'p5' : '              |',
            'p6' : '              |',
            'p7' : '              |',
            'p8' : '      ________|',
            'p3a': '       O      |',
            'p4a': '       |      |',
            'p4b': '       |\     |',
            'p4c': '      /|\     |',
            'p5a': '       |      |',
            'p6a': '        \     |',
            'p6b': '      / \     |',
            'p7a': '   GAME OVER  |'
            }
        if self.lives == 7:
            print('{}\n{}\n{}\n{}\n{}\n{}\n{}\n{}\n'.format(d['p1'], d['p2'], d['p3'], d['p4'], d['p5'], 
                                                            d['p6'], d['p7'], d['p8']))
        if self.lives == 6:
            print('{}\n{}\n{}\n{}\n{}\n{}\n{}\n{}\n'.format(d['p1'], d['p2'], d['p3a'], d['p4'], d['p5'], 
                                                            d['p6'], d['p7'], d['p8']))
        if self.lives == 5:
            print('{}\n{}\n{}\n{}\n{}\n{}\n{}\n{}\n'.format(d['p1'], d['p2'], d['p3a'], d['p4a'], d['p5'], 
                                                            d['p6'], d['p7'], d['p8']))
        if self.lives == 4:
            print('{}\n{}\n{}\n{}\n{}\n{}\n{}\n{}\n'.format(d['p1'], d['p2'], d['p3a'], d['p4b'], d['p5'], 
                                                            d['p6'], d['p7'], d['p8']))
        if self.lives == 3:
            print('{}\n{}\n{}\n{}\n{}\n{}\n{}\n{}\n'.format(d['p1'], d['p2'], d['p3a'], d['p4c'], d['p5'], 
                                                            d['p6'], d['p7'], d['p8']))
        if self.lives == 2:
            print('{}\n{}\n{}\n{}\n{}\n{}\n{}\n{}\n'.format(d['p1'], d['p2'], d['p3a'], d['p4c'], d['p5a'], 
                                                            d['p6'], d['p7'], d['p8']))
        if self.lives == 1:
            print('{}\n{}\n{}\n{}\n{}\n{}\n{}\n{}\n'.format(d['p1'], d['p2'], d['p3a'], d['p4c'], d['p5a'], 
                                                            d['p6a'], d['p7'], d['p8']))
        if self.lives == 0:
            print('{}\n{}\n{}\n{}\n{}\n{}\n{}\n{}\n{}\n'.format(d['p1'], d['p2'], d['p3a'], d['p4c'], d['p5a'], 
                                                            d['p6b'], d['p7'], d['p7a'], d['p8']))
            
# TODO, Create a method for a guess
    def makeGuess(self):
        while self.flag:
            self.ans = ''
            self.ansL = ''
            self.ansU = ''
            if '_' not in self.board:
                print('Congratulations you have guessed {} correctly in {} guess(es)'.format(self.random_word, 8 - self.lives))
                self.playAgain()
                break
            if self.lives == 0:
                print('You have run out of guesses. {} was the word.\n'.format(self.random_word))
                self.playAgain()
                clear_output()
                print('Thanks for playing!')
                break
            self.ans = input("Guess a letter: ")
            self.ansL = self.ans.lower()
            self.ansU = self.ans.upper()
            self.checkGuess()
        return self.flag
            
    def checkGuess(self):
        # check if guess is in it, removes life if not
#         print(self.random_word, self.ans)
        if self.ansU not in self.random_word and self.ansL not in self.random_word:
            clear_output()
            if self.ans in self.guesses:
                print("You have already guessed this letter.")
            if self.ans not in self.guesses:
                self.guesses += self.ans
                self.lives -= 1
            self.printMan()
            print('Guesses left:\t{}\t{}\n Previous guesses:\t{}'.format(self.lives, self.board, self.guesses))
        else:
            clear_output()
            if self.ans in self.guesses:
                print("You have already guessed this letter.")
            if self.ans not in self.guesses:
                self.guesses += self.ans
            self.updateBoard()
            
    # TODO, Create method for updating the board
    def updateBoard(self):
#         clear_output()
        # loops through the random word choosen, if the guess matches changes the nth index in self.toBe to self.ans (Uppercase or Lowercase) 
        for i in range(len(self.random_word)):
            if self.random_word[i] == self.ansU:
                self.toBe[i] = self.ansU
            if self.random_word[i] == self.ansL:
                self.toBe[i] = self.ansL
                
        # checks if it is the first letter or after a space and does the .upper() else it will be .lower()                    
        for i in range(len(self.toBe)):
            if i == 0:
                self.toBe[i] = self.toBe[i].upper()
            elif self.toBe[i-1] == ' ':
                self.toBe[i] = self.toBe[i].upper()
            else:
                self.toBe[i] = self.toBe[i].lower()
                
        # Prints the updated board 
#         print(toBe)
        new_board = ''
        for l in self.toBe:              
            if l == ' ':
                new_board += '  '
            new_board += (l + ' ')
        self.board = new_board
        self.printMan()
        print('Guesses left:\t{}\t{}\n Previous guesses:\t{}'.format(self.lives, self.board, self.guesses))
        self.makeGuess()

    # TODO, Create a method to ask if they want to play again
    def playAgain(self):
        ans = input("Would you like to play again? Y/N ")
        ans.lower()
        if ans == 'y' or ans == 'yes':
            self.lives = 7
            self.startGame()
        if ans != 'y' or ans != 'yes':
            clear_output()
            self.flag = False
            return self.flag
# TODO, Create a while loop to continue until they don't want to play
    def startGame(self):
        while self.flag:
            ans = input("What category would you like to choose from\n 1: Code Terms\n 2: Sports\n ")
            ans = ans.lower()
            if ans == '1' or ans == 'code terms':
                self.makeBoard(ans)
            elif ans == '2' or ans == 'sports':
                self.makeBoard(ans)
            else:
                print('\n\n\tDid you make a typo? Please retry\n\n')
                clear_output()
# TODO(OPTIONAL), Create  a method that adds the body parts on wrong guesses
# TODO(OPTIONAL), Create difficulty levels
h = Hangman()
h.startGame()
```
